```python
import numpy as np
import scipy.signal as signal
import librosa
import soundfile as sf
import librosa.display
import matplotlib.pyplot as plt

# Load the audio file
audio_file, sr = librosa.load('/content/drive/MyDrive/Midterm_DSP/Voices/Galvatron.wav', sr=None)

# Adjust the pitch (n_steps) to control the pitch change
n_steps = -5  # Decrease n_steps
shifted_audio = librosa.effects.pitch_shift(audio_file, n_steps=n_steps, sr=sr)

# Save the processed audio as a WAV file
output_file = '/content/drive/MyDrive/Midterm_DSP/Galvatron_Code.wav'
sf.write(output_file, shifted_audio, sr)

# Print a message to confirm the completion of the processing
print(f"Processed audio saved as {output_file}")

plt.figure(figsize=(10, 4))
librosa.display.waveshow(audio_file, sr=sr)
plt.title('Audio Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import scipy.signal as signal
import librosa
import soundfile as sf
import librosa.display
import matplotlib.pyplot as plt
# Load the audio file
audio_file, sr = librosa.load('/content/drive/MyDrive/Midterm_DSP/Voices/Optimus2.wav', sr=None)


n_steps = -3
shifted_audio = librosa.effects.pitch_shift(audio_file, n_steps=n_steps, sr=sr)

modulated_audio = np.sin(2 * np.pi * 1.5 * np.arange(len(shifted_audio)))
robotic_audio = shifted_audio + 0.3 * modulated_audio


cutoff_hz = 4000
nyquist = 0.5 * sr
normal_cutoff = cutoff_hz / nyquist


b, a = signal.butter(6, normal_cutoff, btype='low', analog=False)


filtered_audio = signal.filtfilt(b, a, robotic_audio)


output_file = '/content/drive/MyDrive/Midterm_DSP/Optimus2_Code.wav'
sf.write(output_file, filtered_audio, sr)

plt.figure(figsize=(10, 4))
librosa.display.waveshow(audio_file , sr=sr)
plt.title('Audio Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```

# Cade Yeager

```python
import numpy as np
import scipy.signal as signal
import librosa
import soundfile as sf

audio_file, sr = librosa.load('/content/drive/MyDrive/Midterm_DSP/Voices/Cade1.wav', sr=None)

n_steps = -2
shifted_audio = librosa.effects.pitch_shift(audio_file, n_steps=n_steps, sr=sr)

time_stretch_factor = 0.9
stretched_audio = signal.resample(shifted_audio, int(len(shifted_audio) * time_stretch_factor))

cutoff_hz = 4000
nyquist = 0.5 * sr
normal_cutoff = cutoff_hz / nyquist

b, a = signal.butter(6, normal_cutoff, btype='low', analog=False)

filtered_audio = signal.filtfilt(b, a, stretched_audio)

sf.write('/content/drive/MyDrive/Midterm_DSP/Voices/Cade1_Code.wav', filtered_audio, sr)
```
